```python
import numpy as np
import pandas as pd

import sqlite3
import warnings
warnings.filterwarnings("ignore")
database = "database.sqlite"
```

```python
netflix = pd.read_csv("./netflix1.csv")
conn = sqlite3.connect(database)
```

```python
netflix.to_sql('netflix_data', conn)
```

Q1	Identify the tables in the dataset and their respective columns.

```python
qry1 = pd.read_sql("""
select * from netflix_data
""", conn)

display(qry1)
```

Q2 -	Determine the number of rows in each table within the schema.

```python
qry2 = pd.read_sql("""
select count(*) as rows from netflix_data
""", conn)

display(qry2)
```

Q3 Identify and handle any missing values in the dataset.

```python
qry3 = pd.read_sql("""
select 
count(case when show_id is null then 1 end) as show_id,
count(case when type is null then 1 end) as type,
count(case when title is null then 1 end) as title,
count(case when director is null then 1 end) as director,
count(case when country is null then 1 end) as country,
count(case when date_added is null then 1 end) as date_added,
count(case when release_year is null then 1 end) as release_year,
count(case when rating is null then 1 end) as rating,
count(case when duration is null then 1 end) as duration,
count(case when listed_in is null then 1 end) as listed_in
from netflix_data

""", conn)

display(qry3)
```

Q4	Analyse the distribution of content types (movies vs. TV shows) in the dataset.

```python
qry4 = pd.read_sql("""
select count(*) as count, type from netflix_data
group by type
""", conn)

display(qry4)
```

Q5 Determine the top 10 countries with the highest number of productions on Netflix.

```python
qry5 = pd.read_sql("""
select count(title) as count, country from netflix_data
group by country
order by count desc
limit 10
""", conn)

display(qry5)
```

Q6 Investigate the trend of content additions over the years.

```python
qry6 = pd.read_sql("""
select count(title) as content_count, release_year from netflix_data
group by release_year
order by release_year
""", conn)

display(qry6)
```

Q7 Analyse the relationship between content duration and release year.

```python
qry7 = pd.read_sql("""
select avg(duration) as avg_duration , release_year, type from netflix_data
group by release_year,type
order by release_year
""", conn)

display(qry7)
```

Q8 Identify the directors with the most content on Netflix.

Q9 Determine the unique genres and categories present in the dataset.

```python
qry9 = pd.read_sql("""
select distinct(listed_in) as genre from netflix_data
""", conn)

display(qry9)
```

Q10 Calculate the percentage of movies and TV shows in each genre.

```python
qry10 = pd.read_sql("""
select * from
( select distinct(listed_in) as abc from netflix_data 
where type = "Movie" ) a
inner join 
( select distinct(listed_in) as abc from netflix_data 
where type = "TV Show" ) b
on a.abc = b.abc
""", conn)

display(qry10)
```

Q11 Identify the most popular genres/categories based on the number of productions.

```python
qry11 = pd.read_sql("""
select count(title) as count, listed_in as genre from netflix_data
group by listed_in
order by count desc
""", conn)

display(qry11)
```

Q12 Calculate the cumulative sum of content duration within each genre.

```python
qry12 = pd.read_sql("""
select listed_in as genre, duration, sum(duration) over(partition by listed_in order by duration) as cumulative_sum
from netflix_data
""", conn)

display(qry12)
```

Q13 Determine the distribution of content releases by month and year

```python
qry13 = pd.read_sql("""
select count(title) as content_count, substr(date_added, 1, instr(date_added, '/') -1) as month, release_year as year
from netflix_data
group by year
""", conn)

display(qry13)
```

Q16 Investigate the distribution of ratings across different genres.

```python
qry16 = pd.read_sql("""
select count(rating) as rat_count, rating, listed_in as genres
from netflix_data
group by rating
""", conn)

display(qry16)
```

```python
qry19 = pd.read_sql("""
SELECT listed_in, COUNT(title) AS title_count
FROM netflix_data
WHERE (LENGTH(listed_in) - LENGTH(REPLACE(listed_in, ',', ''))) = 1
GROUP BY listed_in
HAVING instr(listed_in || ',', ',') > 0;
""", conn)

display(qry19)
```

Q17 Analyse the relationship between ratings and content duration.

Q 18 Identify the most common pairs of genres/categories that occur together in content.

```python
qry18 = pd.read_sql("""
select  listed_in as genre, count(title) as count
from netflix_data
where listed_in like '%,%'
and listed_in not like '%,%,%'
group by listed_in
order by count desc
""", conn)
display(qry18)
```

Q19 Analyse the relationship between genres/categories and content duration.

Q20 Identify the countries where Netflix has expanded its content offerings.
